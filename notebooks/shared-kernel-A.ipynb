```python
# Here we generate two Gaussian distributions

import numpy as np, matplotlib.pyplot as plt, seaborn as sns
sns.set()

x = np.random.normal(size = 30000)
y = np.random.normal(size = 30000)

# this diagram is okay for the research phase
sns.scatterplot(x = x,y = y, color = 'black', s = 1, edgecolor =None, alpha = 1)
```

```python
# For the business report, let's prepare a modified diagram that better fits the audience.

def abracadabra():
    sns.scatterplot(x = x[:3000],y = y[:3000], color = 'black', s = 20, edgecolor ='red', alpha = 1)
```

This workbook A, together with workbook B, demonstrates how to generate management-quality reports directly from Jupyter Notebook.

This workbook contains diagrams together with some code. I don't want to see this code in the final report. I will achieve this with shared kernel. Here is how:

1. This research workbook A serves to define all your variables and methods. Open it in JupyterLab and run it.
2. In the same JupyterLab, open the reporting notebook B in a separate tab. Change the kernel and "steal" the environment from the research notebook A
    (<font color=red>menu Kernel/Change Kernel</font>). Now the notebook B can use the variables and results computed in notebook A.
3. Run the reporting notebook B, so the diagrams get rendered
4. Export the notebook B to HTML or PDF using <font color = red>menu File/Export workbook as </font>.

There is also a way to export workbook B to PDF without showing any source code cells whatsoever, so audience won't even know that abracadabra() was being executed. This can be done at the shell level. This command worked for me at the anaconda prompt:

_jupyter nbconvert notebook_name.ipynb --no-input  --to webpdf --no-prompt --output output_filename_
