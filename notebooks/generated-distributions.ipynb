# some artificially generated distributions

```python
# bonus: generate sample lognormal
import numpy as np, matplotlib.pyplot as plt, seaborn as sns
sns.set()


xn = np.random.normal(size = 10000)
yn = np.random.normal(size = 10000)

xl = np.random.lognormal(size = 10000)
yl = np.random.lognormal(size = 10000)

# y distributions correlated with x
scatter = 0.3
y_xn = [i + np.random.normal()  * scatter for i in xn]
scatter = 1
y_xl = [i + np.random.normal()  * scatter for i in xl]
```

```python
from matplotlib import gridspec
fig =  plt.figure(figsize = (6,10))
gs = gridspec.GridSpec(3,1, height_ratios=[1,1,4]) 
ax0 = plt.subplot(gs[0])
ax1 = plt.subplot(gs[1])
ax2 = plt.subplot(gs[2])


#ax.scatter(x,y, color = 'black', size = 1)
ax = ax.flatten()
axis = ax0
axis.hist(xn, bins = 100)
axis.set_title('x')
axis = ax1
axis.hist(yn, bins = 100)
axis.set_title('y')
axis = ax2
sns.scatterplot(x = xn,y = yn, color = 'black', s = 1, edgecolor = None, ax = axis)
axis.set_title('x vs y')
plt.tight_layout()
```

```python
from matplotlib import gridspec
fig, ax =  plt.subplots(1,3, figsize = (20,6))
ax = ax.flatten()
axis = ax[0]
axis.hist(xn, bins = 100)
axis.set_title('x')
axis = ax[1]
axis.hist(yn, bins = 100)
axis.set_title('y')
axis = ax[2]
sns.scatterplot(x = xn,y = yn, color = 'black', s = 1, edgecolor = None, ax = axis)
axis.set_title('x vs y (uncorrelated)')
plt.tight_layout()
```

```python
from matplotlib import gridspec
fig, ax =  plt.subplots(1,3, figsize = (20,6))
ax = ax.flatten()
axis = ax[0]
axis.hist(xn, bins = 100)
axis.set_title('x')
axis = ax[1]
axis.hist(y_xn, bins = 100)
axis.set_title('y')
axis = ax[2]
sns.scatterplot(x = xn,y = y_xn, color = 'black', s = 1, edgecolor = None, ax = axis)
axis.set_title('x vs y (correlated)')
plt.tight_layout()
```
